    ###### Question 1 (15 points)

    ###### Create a list of numbers which are multiple of 3 and 7 between 1 to 500 (5 points).

```python
# Since 7 & 3 are prime, so multiple of 21 (i.e. 7*3) are our required nos.
w = list(range(21,500,21))
print('List of Numbers (multiples of 3 & 7) is :\n{0}'.format(w))
```

    ###### For the list, find what is the mean. Print what is 25th quartile, 50th quartile and 75th quartile. (10 points)

```python
# Since list is sorted and normalized, so mean & median will coincide. 
c = len(w)
print('Mean of the list is : {0}'.format(sum(w)/c))
print('25th Quartile is : {0}'.format(w[c//4]))
print('50th Quartile is : {0}'.format(w[c//2]))
print('75th Quartile is : {0}'.format(w[3*c//4]))
```

    ###### Question 2 (15 points)

    ###### Read the file ‘ques2.txt’. Print the number of words. (5 points)

```python
## opening file in read mode 
file = open('/Users/ngmodani/Documents/Jupyter Notebook/DataAnalysisBasics/Assignment 1/ques2.txt', 'r') 
file_content = file.read()  ## reading content of the file
words = file_content.split()   ## split the words and counting 
file.close()
print('Count of all words is : {0}'.format(len(words)))
```

    ###### Find out what is the average, median and mode for word length (character length) (10 points).

```python
l_w_len = []  ## create empty list for word length 
for w in words:  ##  iterating over all words 
    l_w_len.append(len(w)) ## storing length of words in the list

## median (middle) element of the sorted list
l_w_len.sort()  ## sorting the list to find median
med_of_len = l_w_len[len(l_w_len)//2] 
print('Median of all word length is : {0}'.format(med_of_len))
## mean (average) of the list
mean_of_len = sum(l_w_len)/len(l_w_len) 
print('Mean of all word length is : {0}'.format(mean_of_len))
## mode (most frequent value) of the list
mod_of_len = max(set(l_w_len), key=l_w_len.count)
print('Mode of all word length is : {0}'.format(mod_of_len))
```

    ###### Question 3 (35 points)

    ###### There are 1500 json files in ‘Lecture03/question03’. Find out how many of the files contain the string, ‘User not found’ (case insensitive). (15 points)
    ###### Create a dictionary, containing (user_id,present). present is 0 if ‘user not found’ else 1. Print random 5 entries (20 points).

```python
import os,json

file_path = '/Users/ngmodani/Documents/Jupyter Notebook/DataAnalysisBasics/Assignment 1/jsons/'
json_files = [f_json for f_json in os.listdir(file_path) if f_json.endswith('.json')]

count=0 ## for count of 'User not found'
D={}    ## for dictionary, containing (user_id,present)
for js in json_files:
    with open(os.path.join(file_path, js)) as data_file:
        data = json.loads(data_file.read())
        if('name' in data and data["name"].lower() == "user not found"):
            count += 1
            D[js.split('.')[0]]=0
        else:
            D[js.split('.')[0]]=1

print("No. files containing the string 'User not found’ (case insensitive) are : {0}".format(count))

##[(key+":"+str(D[key])) for key in list(D.keys())[:5]]
print("\n5 random entries are - ")
for key in list(D.keys())[:5]:
    print("\t"+key+":"+str(D[key]))
```

    ###### Question 4 - 35 Points

    ###### For the same Json files, there is a field, `segmentUid` which can either start with 0xxxx(lets call them 1st party data) or 1xxxxx (lets call them 3rd party data). For each user, count the number of first party segments and third party segments. Points (35 points). Hint : {userID: {first:20,third:30}}

```python
import os,json

file_path = '/Users/ngmodani/Documents/Jupyter Notebook/DataAnalysisBasics/Assignment 1/jsons/'
json_files = [f_json for f_json in os.listdir(file_path) if f_json.endswith('.json')]

segments={}    ## for dictionary, containing {userID: {first:20,third:30}}
for js in json_files:
    count_first=0 ## for count of 0xxxx i.e. 1st party data
    count_third=0 ## for count of 1xxxx i.e. 3rd party data
    with open(os.path.join(file_path, js)) as data_file:
        data = json.loads(data_file.read())
        if('retargetingSegments' in data):
            segments[js.split('.')[0]] = {}
            for retargetingSegment in data['retargetingSegments']:
                if('segmentUid' in retargetingSegment and retargetingSegment['segmentUid'].startswith("0")):
                    count_first += 1
                elif('segmentUid' in retargetingSegment and retargetingSegment['segmentUid'].startswith("1")):
                    count_third += 1
            segments[js.split('.')[0]]['first']= count_first
            segments[js.split('.')[0]]['third']= count_third
print("Following is the number of first party segments and third party segments for each user -\n")   
[(key+ "                       "+str(segments[key])) for key in list(segments.keys())]
```
